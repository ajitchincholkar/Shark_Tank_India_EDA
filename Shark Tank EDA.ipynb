### How much money did each shark invest in season 1 ?

```python
def amount_invested_by_shark(name):
    temp_df = df[df[name.lower()+'_invested'] == 1]
    amount = temp_df['amount_per_shark'].sum()
    print('Amount invested by '+ name.title() + ' is ' + str(round((amount),1)) + ' lakhs ' + '(' + str(round((amount/100),2)) + 'cr).')

sharks=['aman','anupam','ashneer','peyush','namita','vineeta','ghazal']

for shark in sharks:
    amount_invested_by_shark(shark)
```

```python
amount_viz = {'shark':['Aman','Anupam','Ashneer','Peyush','Namita','Vineeta','Ghazal'],'Amount(cr)':[8.95,5.34,4.94,7.7,7.05,3.35,1.3]}
invested_per_shark = pd.DataFrame(amount_viz)
invested_per_shark.sort_values('Amount(cr)',ascending=False,inplace=True)
plt.figure(figsize=(14,6))
plt.bar(x='shark',height='Amount(cr)',data=invested_per_shark)
```
